```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import SelectFromModel
from sklearn.svm import LinearSVC
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import RFE
from warnings import simplefilter
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv(r'C:\Users\elira\DataScience\projects\project_DB_recruit_restaurant_visitor_forecasting\Final\Feature_Selection\2nd\Second_Feature_Selection.csv')
df
```

```python
df_col = df.describe(include ='int64')
df_col.columns
```

```python
varSel = pd.DataFrame({'Variable': df_col.columns[:-1]})
varSel
```

```python
X = df.iloc[:,:-1]
X.head()
```

```python
y = df.iloc[:,-1:]
y.head()
```

### Variable Selection using Random Forest

```python
rfmod = RandomForestClassifier(max_depth=6).fit(X, y.values.ravel())
#rfmod.feature_importances_
```

```python
model = SelectFromModel(rfmod, prefit=True)
model.get_support()
```

```python
varSel['RandomForest'] = model.get_support().astype('int64')
varSel
```

### Summarization and Selection of Variables 


```python
varSel['Sum'] = np.sum(varSel,axis=1)
varSel
```

```python
varSel.groupby('Sum')['Variable'].count()
```

```python
# We can now decide a threshold for selecting our variables!
```

```python
varSel[varSel['Sum']>=1]
```

```python
variables = varSel[varSel['Sum']>=1].iloc[:, [0]]
variables
```

```python
variables_to_columns = variables.transpose().iloc[:].head()
df_variables_to_columns = pd.DataFrame(variables_to_columns)
sum_visitors = ['sum_visitors']
df_variables_to_columns['sum_visitors'] = sum_visitors
df_variables_to_columns.to_csv(r'C:\Users\elira\DataScience\projects\project_DB_recruit_restaurant_visitor_forecasting\Final\Feature_Selection\3rd\third_variables_data.csv',index = False)
df_variables_to_columns
```
